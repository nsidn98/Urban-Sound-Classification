```python
'''
This module is for visualizing the images and verifying
whether some patterns do exist or not in the images
of the same class.
'''

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import math
```

```python
df=pd.read_csv('/Users/siddharthnayak/Downloads/train/train.csv')
```

```python
filepath='/Users/siddharthnayak/Downloads/Urban-Sound-Classification-master/spectograms/'
```

```python
# segregate the dataframe into classes
df_siren=df.loc[df['Class'] == 'siren']
df_street_music=df.loc[df['Class'] == 'street_music']
df_drilling=df.loc[df['Class'] == 'drilling']
df_dog_bark=df.loc[df['Class'] == 'dog_bark']
df_children_playing=df.loc[df['Class'] == 'children_playing']
df_gun_shot=df.loc[df['Class'] == 'gun_shot']
df_engine_idling=df.loc[df['Class'] == 'engine_idling']
df_air_conditioner=df.loc[df['Class'] == 'air_conditioner']
df_jackhammer=df.loc[df['Class'] == 'jackhammer']
df_car_horn=df.loc[df['Class'] == 'car_horn']
```

```python
img=plt.imread('/Users/siddharthnayak/Downloads/Urban-Sound-Classification-master/spectograms/0.spec.png') 
#read the first image of the class
for file in os.listdir(filepath):
    for ID in df_siren.ID:
        if file!='.DS_Store':
            if int(file.split('.')[0])==ID:
                img_new=plt.imread(os.path.join(filepath,file))
                img=((img+img_new)/2)#take mean of the images

plt.imshow(img)
plt.title('Siren')
plt.show()
            
```
